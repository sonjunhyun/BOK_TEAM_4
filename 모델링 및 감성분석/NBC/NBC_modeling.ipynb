```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.naive_bayes import MultinomialNB

# Sample data (대체하여 사용하세요)
# df = pd.DataFrame({'ngram': ['sample ngram data'], 'labels': [1]})

x_data = np.array(df['ngram'])
y_data = np.array(df['labels'])

# Split data
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.1, random_state=7, stratify=y_data)

# Vectorization
vectorizer = CountVectorizer(min_df=15, max_df = 0.95)
x_train_transformed = vectorizer.fit_transform(x_train)
x_test_transformed = vectorizer.transform(x_test)

# Grid search to find best alpha parameter
parameters = {
    'alpha': [0.01, 0.1, 1.0, 10.0, 100.0]
}

grid_search = GridSearchCV(MultinomialNB(), parameters, cv=5, n_jobs=-1)
grid_search.fit(x_train_transformed, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best cross-validation score:", grid_search.best_score_)

# Train with best alpha
best_alpha = grid_search.best_params_['alpha']
model = MultinomialNB(alpha=best_alpha)
model.fit(x_train_transformed, y_train)

print("Training score:", model.score(x_train_transformed, y_train))
print("Test score:", model.score(x_test_transformed, y_test))

# Sample prediction
sample_input = ['구조적 실업자로 전환되면서 자연실업률이 상승']
sample_transformed = vectorizer.transform(sample_input)
y_predict = model.predict(sample_transformed)
print("Predicted label for sample input:", y_predict[0])
```
