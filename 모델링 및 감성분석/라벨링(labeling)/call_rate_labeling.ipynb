#### 채권레포트

```python
df_report = pd.read_csv('/content/drive/Shareddrives/4조/final_data/bond_report_preprocessed.csv')
```

```python
df_report.head()
```

```python
df_report = df_report.rename(columns={'날짜':'date', '텍스트':'text'})
```

```python
df_report.head()
```

```python
df_report['date'] = pd.to_datetime(df_report['date'], format='%y.%m.%d').dt.strftime('%Y-%m-%d')
```

```python
df_report.info()
```

```python
df_report = df_report[~(df_report['date'].str.startswith('2008-') |
                        df_report['date'].str.startswith('2022-') |
                        df_report['date'].str.startswith('2023-'))]
```

```python
df_report
```

```python
df_report['date'].unique
```

```python
call_rate = pd.read_csv('/content/drive/Shareddrives/4조/final_data/labeling_overnight.csv', encoding = 'utf-8')
```

```python
call_rate.tail()
```

```python
call_rate.info()
```

```python
call_rate.isnull().sum()
```

```python
report_call = df_report.merge(call_rate, on='date', how='left')
```

```python
report_call.isnull().sum()
```

```python
report_call.head()
```

```python
report_call.shape
```

```python
report_call = report_call[['date', 'text', 'up_down']]
```
